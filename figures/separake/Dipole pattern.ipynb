```python
%pylab inline
import seaborn as sns
```

```python
d = 2.    # meters
c = 343  # m/s
freqs = [1000, 2000, 3000, 4000]  # Hertz
n_points = 1000
th = np.linspace(0, 360, n_points) * np.pi/180  # angles

# plot parameters
colors = ['r', 'g', 'b', 'k']
colors = sns.light_palette('tangerine', n_colors=9, input='xkcd', reverse=True)
figsize = (3.4, 3.4)

# One image microphone
fig = plt.figure(figsize=figsize)
for i, f in enumerate(freqs):
    fig.add_subplot(2, 2, i+1, polar=True)
    resp = np.abs(1 - np.exp(1j*d*f/c*np.sin(th)))
    plt.polar(th, resp, color=colors[i])
    plt.xticks([])
    plt.yticks([0.5, 1., 1.5, 2., 2.5], [])
    plt.ylim([0., 2.75])
    plt.title(str(f) + ' Hz')
plt.tight_layout(pad=0.5)
plt.savefig('rir_dipoles.pdf', dpi=150)

    
# Two image microphones
fig = plt.figure(figsize=figsize)
for i, f in enumerate(freqs):
    fig.add_subplot(2, 2, i+1, polar=True)
    resp = np.abs(1 - 0.85 * np.exp(1j*d*f/c*np.sin(th)) - 0.85 * np.exp(1j*d*f/c*np.sin(th - np.pi / 2)))
    plt.polar(th, resp, color=colors[i])
    plt.xticks([])
    plt.yticks([0.5, 1., 1.5, 2., 2.5], [])
    plt.ylim([0., 2.75])
    plt.title(str(f) + ' Hz')
plt.tight_layout(pad=0.5)
plt.savefig('rir_tripoles.pdf', dpi=150)
```
