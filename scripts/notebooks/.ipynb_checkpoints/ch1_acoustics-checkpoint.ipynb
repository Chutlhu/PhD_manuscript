```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import pyroomacoustics as pra
from scipy.io import wavfile
import IPython

plt.rcParams.update({'font.size': 14})
```

# Plot RTF to ILD

```python
path_to_audio = '../../media/audio/empress.wav'
fs, audio_anechoic = wavfile.read(path_to_audio)
IPython.display.display(IPython.display.Audio(audio_anechoic, rate=fs))
```

```python
def get_rtf(absorption):

    if absorption=='high':
        absor = pra.Material(0.8, 0.15)
    elif absorption=='medium':
        absor = pra.Material(0.5, 0.15)
    elif absorption=='low':
        absor = pra.Material(0.1, 0.15)
    else:
        raise ValueError("The absorption parameter can only take values ['low', 'medium', 'high']")
        
    pol = np.array([[0,0], [0,4], [3,2], [3,0]]).T
    room = pra.Room.from_corners(pol, fs=fs, max_order=10, materials=absor, ray_tracing=True)

    # Set the ray tracing parameters
    room.set_ray_tracing(receiver_radius=0.5, n_rays=10000, energy_thres=1e-5)
    
    
    # Adding the source
    room.add_source(np.array([1.8, 0.4]), signal=audio_anechoic)

    # Adding the microphones
    R = np.array([[0.5, 0.6],[1.2, 1.8]])
    room.add_microphone_array(pra.MicrophoneArray(R, room.fs))

    # Compute the RIR using the hybrid method
    room.compute_rir()
    
    rir = room.rir
    L = min(len(rir[0][0]), len(rir[1][0]))
    h = np.zeros([L, 2, 1])
    h[:, 0, 0] = rir[0][0][:L]
    h[:, 1, 0] = rir[1][0][:L]
    
    F = 2048

    H1 = np.fft.fft(h[:, 0, 0], F)
    H2 = np.fft.fft(h[:, 1, 0], F)
    Hr = H2 / H1
    
    f = np.linspace(0,fs//2,F//2)
    
    rt60 = room.measure_rt60()

    return f, Hr, h, rt60
```

## RTF

```python
F = 2048
plt.figure(figsize=(24,10))

#### low
f, Hr, hs, rt60 = get_rtf(absorption='high')
rt60 = np.mean(rt60)
L = hs.shape[0]
t = np.arange(L)/fs

plt.subplot(331)
plt.title(r'High Absorption - $\mathtt{RT}_{60} = %1.2f$ seconds'% rt60)
h = hs[:,0,0]/np.max(np.abs(hs[:,0,0]))
plt.plot(t, h, 'k')
plt.ylabel(r'$h_1[n]$')
plt.xlabel('Time [seconds]')
plt.ylim((-0.5, 1.1))

plt.subplot(334)
plt.plot(np.linspace(0,fs//2,F//2), 20*np.log10(np.abs(Hr[:F//2])), 'k')
plt.ylabel(r'$\mathtt{ILD}_{2}[k]$   [dB]')
plt.ylim((-40, 40))
plt.xlabel('Frequency [Hz]')

plt.subplot(337)
plt.plot(np.linspace(0,fs//2,F//2), np.angle(Hr[:F//2]), 'k')
plt.ylabel(r'$\mathtt{IPD}_{2}[k]$   [rad]')
plt.xlabel('Frequency [Hz]')
plt.ylim((-3.5, 3.5))

#### medium
f, Hr, hs, rt60 = get_rtf(absorption='medium')
rt60 = np.mean(rt60)
L = hs.shape[0]
t = np.arange(L)/fs

plt.subplot(332)
plt.title(r'Medium Absorption - $\mathtt{RT}_{60} = %1.2f$ seconds'% rt60)
h = hs[:,0,0]/np.max(np.abs(hs[:,0,0]))
plt.plot(t, h, 'k')
plt.ylabel(r'$h_1[n]$')
plt.xlabel('Time [seconds]')
plt.ylim((-0.5, 1.1))

plt.subplot(335)
plt.plot(np.linspace(0,fs//2,F//2), 20*np.log10(np.abs(Hr[:F//2])), 'k')
plt.ylabel(r'$\mathtt{ILD}_{2}[k]$   [dB]')
plt.ylim((-40, 40))
plt.xlabel('Frequency [Hz]')

plt.subplot(338)
plt.plot(np.linspace(0,fs//2,F//2), np.angle(Hr[:F//2]), 'k')
plt.ylabel(r'$\mathtt{IPD}_{2}[k]$   [rad]')
plt.xlabel('Frequency [Hz]')
plt.ylim((-3.5, 3.5))

#### high
f, Hr, hs, rt60 = get_rtf(absorption='low')
rt60 = np.mean(rt60)
L = hs.shape[0]
t = np.arange(L)/fs

plt.subplot(333)
plt.title(r'Low Absorption - $\mathtt{RT}_{60} = %1.2f$ seconds'% rt60)
h = hs[:,0,0]/np.max(np.abs(hs[:,0,0]))
plt.plot(t, h, 'k')
plt.ylabel(r'$h_1[n]$')
plt.xlabel('Time [seconds]')
plt.ylim((-0.5, 1.1))

plt.subplot(336)
plt.plot(np.linspace(0,fs//2,F//2), 20*np.log10(np.abs(Hr[:F//2])), 'k')
plt.ylabel(r'$\mathtt{ILD}_{2}[k]$   [dB]')
plt.ylim((-40, 40))
plt.xlabel('Frequency [Hz]')

plt.subplot(339)
plt.plot(np.linspace(0,fs//2,F//2), np.angle(Hr[:F//2]), 'k')
plt.ylabel(r'$\mathtt{IPD}_{2}[k]$   [rad]')
plt.xlabel('Frequency [Hz]')
plt.ylim((-3.5, 3.5))

plt.tight_layout()
```
