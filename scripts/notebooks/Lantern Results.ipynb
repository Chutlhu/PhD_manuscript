```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

sns.set()
```

```python
df = df = pd.read_csv("../../assets/lantern/results - Sheet1.csv")
```

```python
curr_df = df.loc[(df['SNR'] == 10) & (df['signal']=='wn+n')]
sns.catplot(x="SNR", y="nRMSE" , hue='Approach', col='$v$',
            kind='bar',
            data=curr_df,
            height=4, aspect=.8)
plt.savefig("snr10.pdf", bbox_inches='tight', dpi=200)
```
