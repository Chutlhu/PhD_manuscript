```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

sns.set()
```

```python
df = df = pd.read_csv("../../assets/lantern/results - Sheet1.csv")
```

```python
curr_df = df.loc[df['signal']=='wn+n']
g = sns.catplot(x="Approach", y="nRMSE" , col='$v$',
                kind='bar',
                data=curr_df,
                height=4, aspect=.8)
g.set_xticklabels(rotation=30)
plt.savefig("snr10.pdf", bbox_inches='tight', dpi=200)
```

```python
df = pd.read_csv("../../assets/lantern/results - Sheet2.csv")
```

```python
curr_df = df.loc[(df['Approach'] != '$GCC-PHAT$') & (df['Approach'] != '$MLP_\mathcal{V}$')]
curr_df = curr_df.loc[df['Approach'] != '$GCC-PHAT$']
sns.catplot(x="SNR", y='nRMSE', hue='Approach', col='$v$',
            kind='bar',
            data=curr_df, height=4, aspect=.8)
plt.savefig("snr.pdf", bbox_inches='tight', dpi=200)
```
